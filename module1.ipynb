```python
import cv2
import matplotlib.pyplot as plt
```

```python
def show_cv2(title, img):
    cv2.imshow(title, img)
    cv2.waitKey()
    cv2.destroyAllWindows()

def show_plt(title, img):
    rgb_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    plt.imshow(rgb_img)
    plt.title(title)
    plt.axis("off")
```

```python
# Tải ảnh màu
color_img = cv2.imread("bgr.png", cv2.IMREAD_COLOR)

plt.figure(figsize=(6,2))
show_plt("color image", color_img)

B, G, R = cv2.split(color_img)

print(f"Blue  - mean: {B.mean():.2f}")
print(f"Green - mean: {G.mean():.2f}")
print(f"Red   - mean: {R.mean():.2f}")
```

```python
# Chuyển ảnh màu thành ảnh xám
gray_img = cv2.cvtColor(color_img, cv2.COLOR_BGR2GRAY)

plt.figure(figsize=(6,2))
show_plt("Color to Gray image", gray_img)
```

```python
# Chuyển ảnh xám sang màu
gray2color_img = cv2.merge([B, G, R])

plt.figure(figsize=(6,2))
show_plt("Gray to Color image", gray2color_img)
```

```python
# Tách kênh màu và biểu diễn dưới dạng ảnh xám
B, G, R = cv2.split(color_img)

plt.figure(figsize=(18,8))
plt.subplot(1,3,1)
show_plt("B", B)
plt.subplot(1,3,2)
show_plt("G", G)
plt.subplot(1,3,3)
show_plt("R", R)
```

```python
# Tái tạo ảnh màu ban đầu
img_reconstructed = cv2.merge([B, G, R])

plt.figure(figsize=(6,2))
show_plt("Reconstructed image", img_reconstructed)
```

```python
# Tạo ảnh mới bằng cách hoán đổi
img_swap_rg = cv2.merge([B, R, G])
img_swap_gb = cv2.merge([G, B, R])
img_swap_br = cv2.merge([R, G, B])

plt.figure(figsize=(12,4))

plt.subplot(2,2,1)
show_plt("Original", color_img)

plt.subplot(2,2,2)
show_plt("Swap R and G", img_swap_rg)

plt.subplot(2,2,3)
show_plt("Swap G and B", img_swap_gb)

plt.subplot(2,2,4)
show_plt("Swap B and R", img_swap_br)
```

```python
# Tạo ảnh mới bằng cách thay thế
img_replace_br = cv2.merge([R, G, R])
img_replace_ggray = cv2.merge([B, gray_img, R])
img_replace_rgray = cv2.merge([B, G, gray_img])

plt.figure(figsize=(12,4))
plt.subplot(1,3,1)
show_plt("img", img_replace_br)
plt.subplot(1,3,2)
show_plt("img", img_replace_ggray)
plt.subplot(1,3,3)
show_plt("img", img_replace_rgray)
```
